```python
df = pd.read_csv('AAPL_daily_11_years_data (2).csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Find samples with missing data
df.isnull().sum()
```

```python
#Delete samples with missing data
df = df.dropna()
```

```python
#Remove Dates and Labels
columns_to_drop =['Date','Increase/Decrease']
features = df.drop(columns_to_drop, axis = 1)
```

```python
#Generate Correlation Heatmap
plt.figure(figsize=(16, 6))
heatmap = sb.heatmap(features.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12)
```
